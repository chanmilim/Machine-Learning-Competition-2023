```python
import matplotlib.pyplot as plt
```

```python
# Now we are going to visualize Histogramm of GR in training data #1

plt.figure(figsize=(5,5))

plt.hist(df.GR.values)
plt.xlabel('GR')
plt.ylabel('Frequency')
plt.title('Gamma-ray record in Train #1')
plt.grid('on')

plt.savefig('figure_1.png')
```

```python
plt.figure(figsize=(3,10))

plt.plot(df.GR.values, df.DEPT.values)
plt.gca().invert_yaxis()
plt.title('GR')
plt.ylabel('DEPTH')
plt.xlabel('Unit: API')
plt.grid('on')

```

```python
plt.figure(figsize=(12,10))
plt.suptitle('Training data #1')

#1
plt.subplot(1, 4, 1)
plt.plot(df.GR.values, df.DEPT.values)
plt.gca().invert_yaxis()
plt.title('Gamma-ray')
plt.ylabel('DEPTH')
plt.xlabel('Unit: API')
plt.grid('on')


#2
plt.subplot(1, 4, 2)
plt.plot(df.RHOB.values, df.DEPT.values)

plt.gca().invert_yaxis()
plt.title('Density')
plt.ylabel('DEPTH')
plt.xlabel('Unit: g/cc')
plt.grid('on')
plt.xlim([2, 2.75])


#3
plt.subplot(1, 4, 3)
plt.plot(df.NPHI.values, df.DEPT.values)

plt.gca().invert_yaxis()
plt.title('Neutron')
plt.ylabel('DEPTH')
plt.xlabel('Unit: %')
plt.grid('on')


#4
plt.subplot(1, 4, 4)
plt.plot(df.RD.values, df.DEPT.values)

plt.gca().invert_yaxis()
plt.title('Resistvity')
plt.ylabel('DEPTH')
plt.xlabel('Unit: m')
plt.xlim([0,25])
plt.grid('on')

plt.tight_layout()
```
